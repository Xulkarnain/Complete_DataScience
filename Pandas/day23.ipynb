## `loc[]` vs `iloc[]` ‚Äî Views vs Copies in Pandas
 When creating subsets of a DataFrame, it's important to know whether you're working with a view (linked to original data) or a copy (independent). This affects whether changes in the subset will also affect the original DataFrame.

üîç `loc[]` (label-based indexing)
- May return a view of the original DataFrame.
- Changes in the subset can reflect in the original data (especially with chained assignments).
- Safer to use with .copy() if you don‚Äôt want that behavior.

```python
subset = df.loc[0:2, ["Name", "Age"]].copy()  # Safe, no link to original

print(subset)
```

üîç `iloc[]` (position-based indexing)
- Usually returns a copy, not a view.
- Modifying the subset does NOT change the original DataFrame.
- Safer for slicing and manipulating without affecting the source.

```python
subset2 = df.iloc[0:2, 0:3]
print(subset2)
```

| Feature        | `loc[]`                          | `iloc[]`                                             |
| -------------- | -------------------------------- | ---------------------------------------------------- |
| Indexing       | Label-based                      | Integer position-based                               |
| Returns        | View or Copy (depends)           | Usually a Copy                                       |
| Risk of change | Might modify original DataFrame  | Safe ‚Äî doesn‚Äôt affect original                       |
| Safety Tip     | Use `.copy()` when using `loc[]` | Generally safe, but `.copy()` is still good practice |

- reindex
---

```python
sales_jan = pd.DataFrame({
    'Sales': [250, 400, 300]
}, index=['Store_A', 'Store_B', 'Store_C'])

sales_jan
```

```python
sales_feb = pd.DataFrame({
    'Sales': [260, 310]
}, index=['Store_A', 'Store_C'])

sales_feb
```

```python
common_index = sales_jan.index
sales_feb_aligned = sales_feb.reindex(common_index) # Aligning sales_feb to sales_jan's index
sales_feb_aligned
```

```python
sales_diff = sales_feb_aligned['Sales'] - sales_jan['Sales']

sales_diff
```

```python
sales_feb_aligned = sales_feb.reindex(common_index, fill_value=0)

sales_feb_aligned
```

```python
sales_diff = sales_feb_aligned['Sales'] - sales_jan['Sales']

sales_diff
```

# Condition based column creation using custom logic
---

```python
# tradtional way to create a new column based on a condition
values = []

for idx, row in df.iterrows():
    if row['Age']< 30:
        values.append('young')
    else:
        values.append('senior')

df['Category'] = values

df
```

- vectorized using apply

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
})
df
```

```python
df['Catogary'] = df['Age'].apply(lambda x : 'young' if x < 30 else 'senior')

df
```
