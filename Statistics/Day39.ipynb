## Skewness tells us whether the data is symmetrical or tilted to one side when plotted on a graph.

### 1. Symmetrical Distribution (No Skew / Normal Distribution)

Mean = Median = Mode

Data is evenly spread on both sides of the center.

Bell-shaped curve.

ðŸ“ˆ Example: Heights of people.

## 2. Positive Skew (Right-Skewed)

Tail is longer on the right side.

Most values are small, but a few very large values pull the mean to the right.

Relationship: Mean > Median > Mode

ðŸ“ˆ Example: Income distribution (a few people earn very high).

## 3. Negative Skew (Left-Skewed)

Tail is longer on the left side.

Most values are high, but a few very small values pull the mean to the left.

Relationship: Mean < Median < Mode

ðŸ“ˆ Example: Age at retirement (most people retire at ~60, but a few much younger).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generate different distributions
normal = np.random.normal(0, 1, 1000)         # Symmetrical
positive = np.random.exponential(1, 1000)     # Right skew
negative = -np.random.exponential(1, 1000)    # Left skew

# Plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Normal
axes[0].hist(normal, bins=30, color="skyblue", edgecolor="black")
axes[0].set_title(f"Normal (Skew={skew(normal):.2f})")

# Positive skew
axes[1].hist(positive, bins=30, color="lightgreen", edgecolor="black")
axes[1].set_title(f"Positive Skew (Skew={skew(positive):.2f})")

# Negative skew
axes[2].hist(negative, bins=30, color="salmon", edgecolor="black")
axes[2].set_title(f"Negative Skew (Skew={skew(negative):.2f})")

plt.show()
```

# Distributions 

## 1. Uniform Distribution

ðŸ‘‰ Every outcome has equal probability.

Example: Tossing a fair die ðŸŽ² â†’ each number 1â€“6 has probability 1/6.

Itâ€™s the simplest distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Uniform distribution between 0 and 1
data = np.random.uniform(0, 1, 10000)

plt.hist(data, bins=30, density=True, color="skyblue", edgecolor="black")
plt.title("Uniform Distribution (0 to 1)")
plt.show()
```

## 2. Bernoulli Distribution

ðŸ‘‰ A distribution with two outcomes: success (1) or failure (0).

Example: Tossing a coin ðŸª™ (Head = 1, Tail = 0).

Controlled by a single parameter p (probability of success).

```python
from scipy.stats import bernoulli

data = bernoulli.rvs(p=0.6, size=10000)  # 1 = success with prob 0.6
plt.hist(data, bins=2, density=True, color="orange", edgecolor="black")
plt.xticks([0,1])
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

## 4. Poisson Distribution

ðŸ‘‰ Counts the number of events in a fixed time/space when events occur independently.

Example: Number of emails you receive per hour.

Controlled by parameter Î» (lambda) = average rate.

```python
from scipy.stats import poisson

mu = 3
x = np.arange(0, 10)
pmf = poisson.pmf(x, mu)

plt.bar(x, pmf, color="purple", edgecolor="black")
plt.title("Poisson Distribution (Î»=3)")
plt.show()
```

## 5. Normal Distribution (Gaussian)

ðŸ‘‰ The most important distribution.

Bell-shaped curve ðŸ›Ž.

Defined by mean (Î¼) and standard deviation (Ïƒ).

Many natural phenomena follow this (height, weight, exam scores).

```python
from scipy.stats import norm

mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color="red")
plt.title("Normal Distribution (Î¼=0, Ïƒ=1)")
plt.show()
```

## Covariance

Meaning: Covariance tells whether two variables move together (positive) or in opposite directions (negative).

Value Range: From -âˆž to +âˆž (not bounded like correlation).




Interpretation:

Positive â†’ Both increase together.

Negative â†’ One increases while the other decreases.

Zero â†’ No linear relationship.

## Correlation

Meaning: Correlation measures the strength and direction of the linear relationship between two variables.

Value Range: -1 to +1

+1 â†’ Perfect positive relation (if X increases, Y increases proportionally).

-1 â†’ Perfect negative relation (if X increases, Y decreases proportionally).

0 â†’ No linear relation.
