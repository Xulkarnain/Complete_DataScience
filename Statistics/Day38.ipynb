```python
# Bayes theorem calculation in Python

# Prior probabilities
P_disease = 0.01          # 1% chance of disease
P_no_disease = 0.99       # 99% no disease

# Likelihoods
P_pos_given_disease = 0.99
P_pos_given_no_disease = 0.05

# Total probability of positive test
P_positive = (P_pos_given_disease * P_disease) + (P_pos_given_no_disease * P_no_disease)

# Posterior probability (Bayes theorem)
P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

print("Probability of having disease given positive test:", round(P_disease_given_positive, 3))
```

üîπ What is PDF?

PDF is used for continuous random variables (not discrete).

It tells us how likely a value is around a specific point.

Unlike PMF (which gives direct probabilities), PDF itself does not give probabilities directly.

Instead, the area under the PDF curve between two points gives the probability.

üëâ In short:

For discrete ‚Üí PMF

For continuous ‚Üí PDF (but you integrate it to get actual probabilities).

üîπ Key Points about PDF

The curve of a PDF always stays above or equal to zero.

The total area under the curve = 1 (because total probability is 1).

Probability of a single exact value (e.g., "height = 170.000 cm") is zero in continuous random variables ‚Äî we only talk about ranges.

Example:

Height of a person = continuous random variable.

PDF shows how likely it is to find someone in a height range, like 160‚Äì170 cm.

```python
!pip install scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mean = 0  
std_dev = 1  

# Generate values
x = np.linspace(-4, 4, 1000)  # from -4 to 4
pdf = norm.pdf(x, mean, std_dev)  # PDF formula applied

# Plot
plt.plot(x, pdf, label="PDF (Normal Distribution)")
plt.fill_between(x, pdf, alpha=0.3)  # shaded area
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (Normal Distribution)")
plt.legend()
plt.show()
```

üîπ PMF (Probability Mass Function) ‚Äì Explained

PMF applies only to discrete random variables (values that are countable, like dice rolls, coin tosses, number of students).

PMF gives the probability of each possible outcome.

Example: Tossing a fair coin ‚Üí PMF assigns P(Head)=0.5, P(Tail)=0.5.

Unlike PDF, PMF does not form a smooth curve but instead bars or points.



```python
import matplotlib.pyplot as plt

# Possible outcomes of a fair die
x = [1, 2, 3, 4, 5, 6]

# Each outcome has equal probability = 1/6
pmf = [1/6] * 6

# Plotting PMF as bar chart
plt.bar(x, pmf, color="skyblue", edgecolor="black")

plt.xlabel("Outcome (x)")
plt.ylabel("P(X = x)")
plt.title("PMF of a Fair Die üé≤")

plt.show()
```

CDF (Cumulative Distribution Function)
1. Simple Explanation

The CDF gives the probability that a random variable 
ùëã
X is less than or equal to some value 
ùë•

In plain words:
üëâ "CDF at point x tells us the total probability accumulated up to that point."

Example:

If you have exam marks (0‚Äì100),

CDF at 40 = Probability(student scores ‚â§ 40).

CDF at 70 = Probability(student scores ‚â§ 70).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Mean and standard deviation of normal distribution
mean, std = 0, 1  

# Generate x values
x = np.linspace(-4, 4, 1000)

# Calculate CDF values
cdf_values = norm.cdf(x, mean, std)

# Plot
plt.plot(x, cdf_values, label="CDF")
plt.title("Cumulative Distribution Function (Normal Distribution)")
plt.xlabel("x")
plt.ylabel("F(x) = P(X ‚â§ x)")
plt.grid(True)
plt.legend()
plt.show()
```
