# Commodity Prices Dataset Analysis

Kaggle Dataset link : https://www.kaggle.com/datasets/debashish311601/commodity-prices

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("data/commodity_futures.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values("Date").reset_index(drop=True)
```

```python
df.head()
```

```python
# Create Year & Month for seasonal/group analysis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

```python
# Quick check of missing values
print(df.isna().sum())
```

```python
# Forward fill missing values
df_ffill = df.ffill()
```

```python
# Check again
print(df_ffill.isna().sum())
```

```python
df_clean = df.ffill().bfill()
```

```python
# Check again
print(df_clean.isna().sum())
```

```python
# Price Trends of Commodities (Descriptive)

plt.figure(figsize=(14,6))
for col in ['ALUMINIUM', 'NICKEL', 'ZINC', 'COTTON', 'GOLD']:
    plt.plot(df_clean['Date'], df_clean[col], label=col)
plt.title("Commodity Price Trends (1990â€“2021)")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```

# Seasonal & Volatility Analysis

```python
import matplotlib.pyplot as plt
```

```python
yearly_avg = df_clean.groupby("Year").mean()
```

```python
# Yearly Average Prices
# See long-term upward/downward movement

plt.figure(figsize=(12,6))
for col in ["GOLD", "WTI CRUDE", "SOYBEANS", "COPPER"]:
    plt.plot(yearly_avg.index, yearly_avg[col], label=col)
plt.title("Yearly Average Commodity Prices")
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Monthly Seasonality
# Are some months consistently higher/lower?

monthly_avg = df_clean.groupby("Month").mean()

plt.figure(figsize=(12,6))
for col in ["GOLD", "WTI CRUDE", "SOYBEANS", "COPPER", "COTTON"]:
    plt.plot(monthly_avg.index, monthly_avg[col], label=col)
plt.title("Monthly Average Commodity Prices (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
# Volatility

returns = df_clean.set_index("Date").pct_change().dropna()
```

```python
print(returns.std())
```

```python
volatility = returns.std().sort_values(ascending=False)
```

```python
volatility
```

```python
plt.figure(figsize=(10,6))
volatility_new = volatility.drop(['Month', 'Year'])
volatility_new.head(10).plot(kind="bar", color="salmon")
plt.title("Top 10 Most Volatile Commodities")
plt.ylabel("Std Dev of Daily Returns")
plt.show()
```
